```python
import matplotlib.pyplot as plt
import numpy as np

pointschl = ([0,309.4], [24,300.7], [48,411.9], [120,328.0], [168,235.2], [216,322.5], [312,315.2], [384,211.2],[480,268.8], [528,221.3], [648,326.6], [696,289.4])
pointsPC = ([0,7.05], [24,7.28], [48,9.84], [120,10.25], [168,7.04], [216,10.13], [312,6.62], [384,15.39],[480,10.94], [528,7.61], [648,13.48], [696,7.28])
pointschlex = ([0,15.43], [24,25.01], [48,25.3], [120,25.3], [168,10.81], [216,19.9], [480,13.53], [528,24.47], [696,13.1])

x1, y1 = zip(*pointschl)
x2, y2 = zip(*pointsPC)
x3, y3 = zip(*pointschlex)

fig, ax1 = plt.subplots(figsize = (12,8))

color = 'tab:red'
ax1.set_xlabel('hours')
ax1.set_ylabel('clorofila in vivo', color=color)
ax1.plot(x1, y1 ,'-ks', label = 'clorofila in vivo')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()

color = 'tab:red'
ax2.set_ylabel('clorofila in vivo', color=color)
ax2.plot(x2, y2, '--k^', label = 'ficocianina in-vivo')
ax2.tick_params(axis='y', labelcolor=color)


ax3 = ax1.twinx()

color = 'tab:red'
ax3.set_ylabel('clorofila extrada', color=color)
ax3.plot(x3, y3, ':k*', label = 'clorofila extrada')
ax3.tick_params(axis='y', labelcolor=color)

fig.tight_layout()

plt.xticks(np.arange(min(x1), max(x1)+1, 24.0))

plt.legend()
```
