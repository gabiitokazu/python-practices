```python
import matplotlib.pyplot as plt
import numpy as np

pointsOD = ([0,7.8], [24,7.4], [48,7.8], [120,5.2], [168,5.2], [216,7.3], [312,5.1], [384,6.0],[480,6.5], [528,12.8], [648,9.9], [696,10.4])
pointsODOD = ([0,19.3], [24,18.9], [48,13.1], [120,12.5], [168,12.8], [216,18.1], [312,13.7], [384,15.2],[480,16.8], [528,33.6], [648,24.5], [696,24.9])

x1, y1 = zip(*pointsOD)
x2, y2 = zip(*pointsODOD)

#plt.plot(x1, y1 ,'-ks', label = 'OD')
#plt.plot(x2, y2, '--k^', label = '% OD')

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)
ax.plot(x1, y1 ,'-ks', label = 'OD')
ax.plot(x2, y2, '--k^', label = '% OD')

plt.xticks(np.arange(min(x1), max(x1)+1, 24.0))
plt.yticks(np.arange(1, 35, 2.5))

plt.legend()

plt.xlim(0,696)

fig.savefig('AlgalTraps_OD_ODOD.png', dpi = 300)
```
